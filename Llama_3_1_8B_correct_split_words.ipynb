```python
import transformers
import torch
import time

# Start the timer
start_time = time.time()

model_id = "meta-llama/Meta-Llama-3.1-8B-Instruct"

pipeline = transformers.pipeline(
    "text-generation",
    model=model_id,
    model_kwargs={"torch_dtype": torch.bfloat16},
    device_map="auto",
)

messages = [
    {"role": "system", "content": "You are a helpful assistant!"},
    {"role": "user", "content": "Correct the mistakenly seperated words in the following text: Lo Stato e la Chiesa ca ttolica sono molto importanti nella storia italiana. Puo capitare che questi errori si ve rifichino."},
]

outputs = pipeline(
    messages,
    max_new_tokens=256,
)

# Stop the timer
end_time = time.time()

# Calculate the duration in minutes
duration_minutes = (end_time - start_time) / 60

print(outputs[0]["generated_text"][-1])
print(f"Inference took {duration_minutes:.2f} minutes.")
```
